```python
from tensorflow.keras.models import load_model
import cv2
from matplotlib import pyplot as plt
import numpy as np
import tensorflow as tf
```

```python
loaded_model = load_model('Model//best_model.h5')
```

```python
img = cv2.imread('C://Projects//Malaria Cell Prediction using CNN//Dataset//cell_images//cell_images//Uninfected//C241NThinF_IMG_20151207_124608_cell_40.png')
plt.imshow(img)
plt.show()
```

```python
resize = tf.image.resize(img, (256,256))
plt.imshow(resize.numpy().astype(int))
plt.show()
```

```python
result = loaded_model.predict(np.expand_dims(resize/255, 0))
```

```python
if result > 0.5:
    print("Non Parasitized")
else:
    print("Parasitized")
```
